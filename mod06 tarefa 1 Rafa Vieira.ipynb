####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python

# Define o tamanho geral dos gráficos
plt.figure(figsize=(15, 10))

variaveis = ["APGAR1", "APGAR5", "SEXO", "GESTACAO", "ESTCIVMAE"]

# Cria um subplot para cada variável
for i, var in enumerate(variaveis, 1):
    plt.subplot(2, 3, i)  # Organiza os gráficos em uma grade
    sns.countplot(data=df, x=var, palette="viridis")
    plt.title(f"Distribuição de {var}")
    plt.xticks(rotation=45)  # Rotaciona os rótulos no eixo X para melhor leitura

plt.tight_layout()  # Ajusta o espaçamento entre os gráficos
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Define o tamanho do gráfico
plt.figure(figsize=(10, 6))

# Cria o gráfico de barras com a média de APGAR5 por CONSULTAS
sns.barplot(data=df, x="CONSULTAS", y="APGAR5", ci=None, palette="magma")

# Adicionar título e rótulos
plt.title("Média do APGAR5 por Número de Consultas Pré-natais")
plt.xlabel("Número de Consultas Pré-natais")
plt.ylabel("Média do APGAR5")

# Exibe o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Define o tamanho da figura
plt.figure(figsize=(15, 10))

# Lista das variáveis para plotar
variaveis = ["SEXO", "GESTACAO", "ESTCIVMAE", "CONSULTAS"]

# Cria um subplot para cada variável
for i, var in enumerate(variaveis, 1):
    plt.subplot(2, 2, i)  # Organiza os gráficos em uma grade 2x2
    sns.countplot(data=df, x=var, palette="coolwarm")
    plt.title(f"Frequência de {var}")
    plt.xticks(rotation=45)  # Rotaciona os rótulos para melhor visualização

plt.tight_layout()
plt.show()
```
